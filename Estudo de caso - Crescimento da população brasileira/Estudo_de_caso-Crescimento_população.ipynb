```python
import matplotlib.pyplot as plt
```

```python
#DataSus para adquirir os dados


dados = open("populacao_brasileira.csv").readlines()
```

```python
x = []
y = []

for i in range(len(dados)):#len conta quantas linhas tem, range cria uma lista q vai de 0 até o valor de len
    if i != 0:
        linha = dados[i].split(";") #quebrar as linhas do csv
        x.append(int(linha[0]))
        y.append(int(linha[1]))
```

Começa mostrando aos poucos, o código abaixo já mostra o gráfico na versão final

```python
plt.bar(x,y, color = "#e4e4e4")
plt.plot(x,y, color = 'k', linestyle = '--')
plt.title("Crescimento da população brasileira de 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x 100.000.000")
#plt.show()
plt.savefig("População_brasileira.png", dpi = 300)
```
