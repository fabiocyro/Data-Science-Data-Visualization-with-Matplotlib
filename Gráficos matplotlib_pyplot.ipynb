## Visualização de dados com matplotlib.pyplot, bem introdutório (mas dá certificado hehehehe)

```python
import matplotlib.pyplot as plt
```

Tudo será praticamente feito na linha a seguir. A aula funciona como uma evolução, tipo na primera aula x e y só possuíam 2 valores, depois aumentou, depois colocou título e por aí vai

## Linhas

```python
x = [1, 2, 5]
y = [3, 4, 7]

titulo = "Gráfico de linhas"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x,y)
plt.show()
```

## Barras

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de barras"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x,y)
plt.show()
```

## Comparando barras

```python
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]

titulo = "Gráfico de barras 2"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1,y1, label = "Grupo 1")
plt.bar(x2,y2, label = "Grupo 2")
plt.legend()

plt.show()
```

## Scatter


```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de dispersão ou Scatterplot"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label = "Meus pontos", color = 'r') #Plota apenas os pontos
plt.plot(x,y) #PLota linhas ligando
plt.legend() #Esse comando exibe todas as label
plt.show()
```

## Configurando o gráfico

Aqui é mostrando como mudar cor, tipo de linha etc.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de dispersão ou Scatterplot"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label = "Meus pontos", color = 'r', marker = '.', s = 200) 
plt.plot(x,y, color = 'k', linestyle = ':') 
plt.legend() 
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de dispersão ou Scatterplot"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

#USANDO CORES EM MODO HEXADECIMAL (Recomendado pois é só ir na net e pesquisar o valor que deseja)
plt.scatter(x,y, label = "Meus pontos", color = 'r', marker = '.', s = 200) 
plt.plot(x,y, color = '#009900', linestyle = '-') 
plt.legend() 
plt.show()
```

```python
# COLOCANDO TAMANHOS DIFERENTES PARA OS PONTOS DO SCATTER
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
z = [20, 5, 100, 33, 10]
titulo = "Gráfico de dispersão ou Scatterplot"
eixox = "Eixo x"
eixoy = "Eixo y"

#Título
plt.title(titulo)

#Eixos

plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label = "Meus pontos", color = 'r', marker = '.', s = z) #muda-se aqui
plt.plot(x,y, color = '#009900', linestyle = '-') 
plt.legend() 
plt.show()
```
